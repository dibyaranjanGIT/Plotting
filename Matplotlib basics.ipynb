```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.title('Simple Graph')
plt.xlabel('prices')


plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.title('Simple Graph')
plt.xlabel('prices')
plt.xticks([1,2,3,])
plt.yticks([1,2,3,4,5,6]) 

plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y,label='2x',color='red',linewidth=2,marker='*',markersize=5)
plt.title('Simple Graph')
plt.xlabel('prices')
plt.xticks([1,2,3,])
plt.yticks([1,2,3,4,5,6]) 
plt.legend()

plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y,label='2x',color='red',linewidth=2,marker='*',markersize=5)
plt.title('Simple Graph')
plt.xlabel('prices')
plt.xticks([1,2,3,])
plt.yticks([1,2,3,4,5,6]) 
plt.legend()

x2 = np.arange(0,4,1)
y2 = x2**2

plt.plot(x2,y2,label='x*2')
plt.legend()


plt.show()
```

```python
plt.figure(figsize=(10,6))

x = [1,2,3]
y = [2,4,6]

plt.plot(x,y,label='2x',color='red',linewidth=2,marker='*',markersize=5)
plt.title('Simple Graph')
plt.xlabel('prices')
plt.xticks([1,2,3,])
plt.yticks([1,2,3,4,5,6]) 
plt.legend()

x2 = np.arange(0,4,1)
y2 = x2**2

plt.plot(x2,y2,label='x*2')
plt.legend()


plt.show()

# To save the figure
plt.savefig('pltgraph.png')
```

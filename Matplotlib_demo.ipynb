```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')

gas
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))


plt.plot(gas.Year,gas.USA,label='USA',marker='*')
plt.plot(gas.Year,gas.Canada,label='Canada',marker='*')
plt.plot(gas.Year,gas['South Korea'],label='SK',marker='*')

plt.xticks(gas.Year[::3])
plt.ylabel("Prices of gas in $")

plt.title("Gas prices over time")
plt.legend()

plt.savefig('gas_price.png')
 
plt.show()
```

```python
left = fifa[fifa['Preferred Foot'] == 'Left'].count()[0]
right= fifa[fifa['Preferred Foot'] == 'Right'].count()[0]
```
